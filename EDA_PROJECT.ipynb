```python
import pandas as pd
import matplotlib .pyplot as plt
```

```python
data=pd.read_csv('StudentsPerformance.csv')
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
grouped = df.groupby('math score')['reading score'].mean().reset_index()
plt.figure(figsize=(10,5))
plt.bar(grouped['math score'], grouped['reading score'])
plt.title("Average Reading Score for Each Math Score")
plt.xlabel("Math Score")
plt.ylabel("Average Reading Score")
plt.xticks(rotation=45)
plt.show()
```

```python
grouped = df.groupby('math score')['reading score'].mean().reset_index()
plt.figure(figsize=(8,5))
plt.scatter(grouped['math score'], grouped['reading score'])
plt.title("Math Score vs Average Reading Score")
plt.xlabel("Math Score")
plt.ylabel("Average Reading Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
grouped = df.groupby('math score')['reading score'].mean().reset_index()
plt.plot(grouped['math score'], grouped['reading score'])
plt.title("Math Score vs Average Reading Score")
plt.xlabel("Math Score")
plt.ylabel("Average Reading Score")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
gender_distribution = df['gender'].value_counts()
plt.title("Gender Distribution")
plt.pie(gender_distribution, labels=gender_distribution.index, autopct='%1.1f%%',startangle=90)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['math score'], bins=10, alpha=0.5, label='Math Score', color='skyblue')
plt.hist(df['reading score'], bins=10, alpha=0.5, label='Reading Score', color='lightgreen')
plt.hist(df['writing score'], bins=10, alpha=0.5, label='Writing Score', color='salmon')
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Math, Reading, and Writing Scores")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
grouped = df.groupby('math score')['reading score'].mean().reset_index()
gender_grouped = df.groupby('gender').size()
plt.figure(figsize=(14,10))

# Bar Chart
plt.subplot(2,3,1)
plt.bar(grouped['math score'], grouped['reading score'])
plt.xlabel("Math Score")
plt.ylabel("Reading Score")

# Scatter Chart
plt.subplot(2,3,2)
plt.scatter(grouped['math score'], grouped['reading score'])
plt.xlabel("Math Score")
plt.ylabel("Reading Score")

# Line Chart
plt.subplot(2,3,3)
plt.plot(grouped['math score'], grouped['reading score'])
plt.xlabel("Math Score")
plt.ylabel("Reading Score")

# Pie Chart (Gender Distribution)
plt.subplot(2,3,4)
plt.pie(
    gender_grouped,
    labels=gender_grouped.index,
    autopct='%1.1f%%',
    startangle=90
)

# Histogram
plt.figure(figsize=(8,5))
plt.hist(df['math score'], bins=10, alpha=0.5, label='Math Score', color='skyblue')
plt.hist(df['reading score'], bins=10, alpha=0.5, label='Reading Score', color='lightgreen')
plt.hist(df['writing score'], bins=10, alpha=0.5, label='Writing Score', color='salmon')
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.title("Distribution of Math, Reading, and Writing Scores")
plt.legend()
plt.show()

plt.tight_layout()
plt.show()
```
